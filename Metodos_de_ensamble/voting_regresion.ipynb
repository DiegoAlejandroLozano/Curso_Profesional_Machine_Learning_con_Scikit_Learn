```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.neighbors import KNeighborsRegressor

from matplotlib import pyplot as plt
```

```python
#Creando los modelos
modelo1 = LinearRegression()
modelo2 = RandomForestRegressor(n_estimators=10, random_state=1)
modelo3 = KNeighborsRegressor()
```

Experimento 1

```python
#Creando los datos
X = np.array([
    [1,1],
    [2,4],
    [3,9],
    [4,16],
    [5,25],
    [6,36]
])

y = np.array([2,6,12,20,30,42])
```

```python
#Creando el ensamble tipo voting para regresión lineal
lista_modelos = [
    ('modelo1', modelo1),
    ('modelo2', modelo2),
    ('modelo3', modelo3)
]
ensamble = VotingRegressor(estimators=lista_modelos)
print('La predicción es: {0}'.format(ensamble.fit(X,y).predict(X)))
print('puntaje es: {puntaje:.4f}%'.format(puntaje=ensamble.score(X,y)))
```

Experimento 2

```python
#Se generan los datos
x = np.linspace(0,100,num=1000)
y = 10*np.sin(x)
```

```python
#Se grafican los datos
plt.figure(figsize=(15,6))
plt.plot(x,y)
plt.xlim((0,100))
plt.ylim((-15,15))
plt.grid()
```

```python
#Se entrena el modelos con los nuevos datos
ensamble.fit(x.reshape(-1,1), y)
y_pred = ensamble.predict(x.reshape(-1,1))
```

```python
#Graficando los valores reales y los valores que se han predecido 
plt.figure(figsize=(15,6))
plt.plot(x,y, label='Valores reales')
plt.plot(x,y_pred, label='Valores predicción')
plt.grid()
plt.legend()
plt.xlim((0,100))
plt.ylim((-15,15))
plt.show()
```

```python
print('La puntuación es: {puntuacion:.4f}%'.format(puntuacion = ensamble.score(x.reshape(-1,1), y)*100))
```

Experimento 3

```python
#Se generan los datos
x3 = np.linspace(0,5,num=1000)
y3 = np.exp(x3)
```

```python
#Se grafican los datos
plt.figure(figsize=(15,6))
plt.plot(x3,y3)
plt.grid()
```

```python
#Se entrena el modelos con los nuevos datos
ensamble.fit(x3.reshape(-1,1), y3)
y3_pred = ensamble.predict(x3.reshape(-1,1))
```

```python
print('La puntuación es: {puntuacion:.4f}%'.format(puntuacion = ensamble.score(x3.reshape(-1,1), y3)*100))
```

```python
#Graficando los valores reales y los valores que se han predecido 
plt.figure(figsize=(15,6))
plt.plot(x3,y3, label='Valores reales')
plt.plot(x3,y3_pred, label='Valores predicción')
plt.grid()
plt.legend()
plt.show()
```

```python
x3_real = 3
y3_real = np.exp(x3_real)
y3_pre = ensamble.predict([[x3_real]])
print('x_real = {x3_r} y_real = {y3_r:.3f}'.format(x3_r=x3_real, y3_r=y3_real))
print('x_real = {x3_r} y_real = {y3_p:.3f}'.format(x3_r=x3_real, y3_p=y3_pre[0]))
```
