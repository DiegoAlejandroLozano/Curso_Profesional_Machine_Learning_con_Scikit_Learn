```python
#Importando las librerias
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
```

```python
#Cargando los datos
dataset = pd.read_csv('..\Datasets\candy.csv')
dataset.head(10)
```

```python
X = dataset.drop('competitorname', axis=1)
X
```

```python
#Creando el modelo
kmeans = MiniBatchKMeans(n_clusters=4, batch_size=8).fit(X)
```

```python
#Verificando los resultados
print('Total de centros: {}'.format(len(kmeans.cluster_centers_)))
print('='*64)
print(kmeans.predict(X))
```

```python
#Uniendo las grupos con el dataset original
dataset['grupos'] = kmeans.predict(X)
dataset
```
