```python
import pandas as pd
from sklearn.cluster import MeanShift
```

```python
#Importando los datos
dataset = pd.read_csv('..\Datasets\candy.csv')
dataset.head()
```

```python
#Se elimina la columna competitorname ya que es una columna categórica que no se puede entrenar con el algoritmo
X = dataset.drop('competitorname', axis=1)
X.head()
```

```python
#Creando y entrenando el modelo
meanshift = MeanShift().fit(X)
```

```python
#Se verifica la cantidad de grupos creados
print('La cantidad de grupos creados es {}'.format(max(meanshift.labels_)+1))
```

```python
#Integrandos las etiquetas encontradas por el algorítmo al dataset
dataset['meanshift'] = meanshift.labels_
dataset
```
